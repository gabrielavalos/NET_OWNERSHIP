```python
#SUM PRODUCTION FOR EACH DATE FOR EACH PARTNER
cml_oil = pd.Series(nets_df.groupby("Date")["CML-OIL"].sum())
cml_gas = pd.Series(nets_df.groupby("Date")["CML-GAS"].sum())
ken_oil = pd.Series(nets_df.groupby("Date")["KEN-OIL"].sum())
ken_gas = pd.Series(nets_df.groupby("Date")["KEN-GAS"].sum())
casey_oil = pd.Series(nets_df.groupby("Date")["CASEY-OIL"].sum())
casey_gas = pd.Series(nets_df.groupby("Date")["CASEY-GAS"].sum())
phil_oil = pd.Series(nets_df.groupby("Date")["PHIL-OIL"].sum())
phil_gas = pd.Series(nets_df.groupby("Date")["PHIL-GAS"].sum())
idc_oil = pd.Series(nets_df.groupby("Date")["IDC-OIL"].sum())
idc_gas = pd.Series(nets_df.groupby("Date")["IDC-GAS"].sum())
jason_oil = pd.Series(nets_df.groupby("Date")["JASON-OIL"].sum())
jason_gas = pd.Series(nets_df.groupby("Date")["JASON-GAS"].sum())
lanmln_oil = pd.Series(nets_df.groupby("Date")["LANMLN-OIL"].sum())
lanmln_gas = pd.Series(nets_df.groupby("Date")["LANMLN-GAS"].sum())
patteron_oil = pd.Series(nets_df.groupby("Date")["PATTERSON-OIL"].sum())
patterson_gas = pd.Series(nets_df.groupby("Date")["PATTERSON-GAS"].sum())
travis_oil = pd.Series(nets_df.groupby("Date")["TRAVIS-OIL"].sum())
travis_gas = pd.Series(nets_df.groupby("Date")["TRAVIS-GAS"].sum())
```

```python
#PUT ALL DAILY SUMMED DATA IN SINGLE DF WITH MULTI COLUMNS (PARTNER, OIL & GAS)
#DATA GOING INTO DF (DATE IS COLUMN NAME IS DATA WAS GROUPED BY DATE)
data= [cml_oil, cml_gas, ken_oil,ken_gas,casey_oil,casey_gas,phil_oil,phil_gas,idc_oil,idc_gas ,jason_oil,jason_gas ,lanmln_oil,lanmln_gas,patteron_oil,patterson_gas,travis_oil,travis_gas]
#CREATE DATAFRAME, TRANSPOSE SO DATE IS IN ROW POSITION
df = pd.DataFrame(data).transpose()
#NEW COLUMN NAMES CREATED BASED ON PARTNER AND THE OWNERSHIIP (OIL & GAS)
partners_output = [["CML", "Ken", "Casey", "Phil", "IDC", "Jason", "LANMLN", "Patterson", "Travis"],["OIL", "GAS"]]
#MAXING COLUMN NAMES FIT MULTI-INDEX FORMAT USING COLUMN NAMES CREATED IN ABOVE LINE
df.columns = pd.MultiIndex.from_product(partners_output, names=["partner","ownership"])
#RESETING INDEX SO THAT DATE COLUMN CAN BE CONVERTED TO DATETIME AND SORTED
df.reset_index(inplace=True)
```
